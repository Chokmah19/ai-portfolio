Pandas實作練習

````python
import pandas as pd

# 讀取 CSV
train_df = pd.read_csv('titanic_data/train.csv')
test_df  = pd.read_csv('titanic_data/test.csv')
````

**※初學者易混淆的內容**

| 表達式               | 說明                                           |
|----------------------|----------------------------------------------|
| `df['欄位名稱']`      | ✅ 標準做法，用來**取出一個欄位**（Series）       |
| `df[['A', 'B']]`      | ✅ 取出多個欄位（傳入一個 list）                |
| `df['A', 'B']`        | ❌ 不可用 tuple，會報錯                         |
| `df('欄位名稱')`       | ❌ 如果 `df` 是 DataFrame，不能用小括號存取欄位    |
| `func()`             | ✅ 小括號是**函數的呼叫**，例如 `sum()`、`mean()` |

小結：
- **取欄位要用中括號 []，因為 DataFrame 是用 dict-like 結構設計的。**
- 小括號 () 是函數用的，例如：

```python
print()、sum()、df.mean()
```

````python
# 1. 查看資料維度
print("訓練集：", train_df.shape)
print("測試集：", test_df.shape)

# 2. 前五筆與後五筆
display(train_df.head())
display(train_df.tail())

# 3. 欄位資訊與缺失值統計
print(train_df.info())
print(train_df.isnull().sum())

# 4. 基本統計描述
display(train_df.describe())

# 5. 單變量分析：檢視年齡分布
train_df['Age'].hist(bins=30)

# 6. 依性別分組，計算每組平均存活率
survival_by_sex = train_df.groupby('Sex')['Survived'].mean()
print(survival_by_sex)

# 7. 補齊年齡缺失值：用中位數
age_median = train_df['Age'].median()
train_df['Age'].fillna(age_median,inplace=True)
````

````python
# 8. 新增一個新資料欄，代表乘客的『艙等（Pclass）』與『性別（Sex）』的組合，用來觀察是否能提供額外的資訊，協助預測是否生還。
# 將性別文字轉換成數字，新增一個「性別編碼」資料欄
train_df['Sex_Code'] = train_df['Sex'].map({'male': 0, 'female': 1})
# 建立新特徵：艙等 * 性別編碼
train_df['Pclass_Sex'] = train_df['Pclass'] * train_df['Sex_Code']

display(train_df[['Pclass','Sex','Pclass_Sex']].head())
````

🧠背後的想法  
在機器學習裡，我們常會創造**新的特徵欄位（feature columns）**，希望能幫助模型更有效率地學習、分類、預測。

這裡的思路是：

- 「艙等」（Pclass）代表乘客付的票價等級，數字越小越高級（1 是最高，3 是最低）。

- 「性別」（Sex）可能跟生還機率有關（例如女性或兒童被優先救援）。

- 如果把這兩個變數合併起來，我們可能能看出：
👉「低艙等女性」與「高艙等男性」的生還機率是否不同。
